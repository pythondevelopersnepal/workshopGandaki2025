```python
import numpy as np

# Simulate a small 2x2 RGB image
image = np.array([
    [[255, 0, 0], [0, 255, 0]],
    [[0, 0, 255], [255, 255, 0]]
])  # Shape: (2, 2, 3)

# Flatten it
vector = image.flatten()
print("Flattened vector:", vector)
print("Shape:", vector.shape)
```

### live example

```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load image and resize it to manageable size (e.g., 128x128)
img = load_img('earthquake.png', target_size=(128, 128))  # Resize to (128x128)

# Show the image
plt.imshow(img)
plt.axis('off')
plt.title('Original Earthquake Image')
plt.show()
```

```python
# Convert image to NumPy array
img_array = img_to_array(img)

print("Image Shape (H, W, C):", img_array.shape)
```

```python
img_array = img_array / 255.0  # Normalize pixel values to [0, 1]
```

```python
img_vector = img_array.flatten()
print("Flattened Vector Shape:", img_vector.shape)
```
