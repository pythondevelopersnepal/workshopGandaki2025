# S1.2_pandas_matplotlib

- When you run this code, it will:

- Download the Iris dataset (150 rows, 5 columns)
- Show a preview of the data
- Calculate average measurements for each species
- Create a scatter plot of sepal measurements
- Display basic statistical summaries

- Library: pip install pandas matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
from statistics import mean
```

```python
# Loading data from a URL
url = "https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv"
iris_df = pd.read_csv(url)
```

```python
print("Dataset Preview:")
print(iris_df.head())  # Shows first 5 rows
print("\nDataset Shape:", iris_df.shape)  # Shows dimensions
```

```python
# List comprehension to get average measurements per species
species_means = {
    species: [mean(iris_df[iris_df['species'] == species][col]) 
             for col in ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
    for species in iris_df['species'].unique()
}
```

```python
# Convert to DataFrame for better display 
means_df = pd.DataFrame(species_means, 
                       index=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
print("\nAverage measurements by species:")
print(means_df)
```

```python
# Simple visualization 
plt.figure(figsize=(10, 6))
for species in iris_df['species'].unique():
    species_data = iris_df[iris_df['species'] == species]
    plt.scatter(species_data['sepal_length'], 
                species_data['sepal_width'], 
                label=species)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Iris Sepal Measurements by Species')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Basic statistical summary 
print("\nBasic Statistics:")
print(iris_df.describe())
```
