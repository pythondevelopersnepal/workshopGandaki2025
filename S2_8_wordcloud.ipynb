```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# List of 50 words related to AQI and air pollution
words = [
    "AQI", "air pollution", "smog", "particulate matter", "PM2.5", 
    "PM10", "ozone", "nitrogen dioxide", "sulfur dioxide", "carbon monoxide",
    "Nepal", "Kathmandu", "most polluted city", "Pokhara", "haze",
    "emissions", "vehicle exhaust", "industrial pollution", "dust", "smoke",
    "health risks", "respiratory issues", "asthma", "lung disease", "toxic air",
    "air quality", "monitoring", "pollutants", "urban", "rural",
    "forest fires", "construction dust", "weather", "wind", "humidity",
    "temperature", "fossil fuels", "coal burning", "factories", "traffic",
    "mask", "purifier", "environment", "climate", "exposure",
    "contamination", "visibility", "seasonal", "public health"
]

# Join words into a single string
text = " ".join(words)

# Create word cloud
wordcloud = WordCloud(width=800, height=400,
                      background_color='white',
                      min_font_size=10,
                      max_words=50,
                      collocations=False).generate(text)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('AQI and Air Pollution Word Cloud')
plt.show()
```
